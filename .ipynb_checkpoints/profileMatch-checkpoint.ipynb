```python
def getLikelyProfile(userData):
    

    sum = 0
    max = -1
    match = dict()
    index = 0
    j = 0
    profileNames = ['Calcium deficiency', 'Phosphorus deficiency', 'Nitrogen deficiency', 'Potassium deficiency', 'Boron deficiency', 'Copper deficiency', 'Fungus gnats', 'Heat Stress', 'Iron deficiency', 'Light Burn', 'Magnesium deficiency', 'Mildew', 'Nute burn', 'Overwatering', 'pH fluctuation', 'Root rot', 'Spider mites', 'Underwatering', 'Worm/ caterpillar Infestation']
    probabilities = {'Calcium deficiency': 0, 'Phosphorus deficiency': 0, 'Nitrogen deficiency': 0, 'Potassium deficiency': 0, 'Boron deficiency': 0, 'Copper deficiency': 0, 'Fungus gnats': 0, 'Heat Stress': 0, 'Iron deficiency': 0, 'Light Burn': 0, 'Magnesium deficiency': 0, 'Mildew': 0, 'Nute burn': 0, 'Overwatering': 0, 'pH fluctuation': 0, 'Root rot': 0, 'Spider mites': 0, 'Underwatering': 0, 'Worm/ caterpillar Infestation': 0}
    
    for profile in profile_list:
        
        for symptom in profile:
            i = 0
            for i in range(0, len(profile[symptom])):
                if userData[symptom] == str(profile[symptom][i]):
                    sum += 1
                elif userData[symptom] != "no value":
                    sum = sum - 1
                
        probabilities[profileNames[j]] = sum/len(profile)
            
        if (sum > 0 and sum/len(profile) > max):
            match = profile
            max = sum/len(match)
        
        j += 1
        sum = 0  
     
    for profile in profile_list:
        if match == profile:
            break
        index += 1
        
    if match != {}:
        return (profileNames[index], str(max*100), match, probabilities)
    else:
        return -1
```
